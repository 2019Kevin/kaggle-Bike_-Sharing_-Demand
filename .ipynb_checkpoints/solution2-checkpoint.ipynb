# 特征工程

## 根据datetime列创建新的特征列


```python
dailyData["date"] = dailyData.datetime.apply(lambda x: x.split()[0])
dailyData["hour"] = dailyData.datetime.apply(lambda x: x.split()[1].split(":")[0])
dailyData["weekday"] = dailyData.date.apply(lambda dateString: 
                                            calendar.day_name[datetime.strptime(dateString, "%Y-%m-%d").weekday()])
dailyData["month"] = dailyData.date.apply(lambda dateString: 
                                          calendar.month_name[datetime.strptime(dateString, "%Y-%m-%d").month])

# 将season和weather根据数据描述进行转换, 这两个特征是类别变量, 先将相应的数值转换为字符串, 以便后续的可视化操作
dailyData["season"] = dailyData.season.map({1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"})
dailyData["weather"] = dailyData.weather.map({
        1: "Clear, Few clouds, Partly cloudy, Partly cloudy",
        2: "Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist",
        3: "Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds",
        4: "Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog"
    })
```

```python
# 强制转换到category类别类型

categoryVariable = ["season", "holiday", "workingday", "weather", "hour", "weekday", "month"]
for var in categoryVariable:
    dailyData[var] = dailyData[var].astype("category")
```

```python
# 去掉不需要的特征列
dailyData = dailyData.drop(["datetime"], axis=1)
```

```python
# 变量的数据类型的简单可视化

dataTypeDf = pd.DataFrame(dailyData.dtypes.value_counts()).reset_index().rename(columns={"index": "variableType", 0: "count"})
fig, ax = plt.subplots()
fig.set_size_inches(12, 5)
# 画出条形图
sns.barplot(data=dataTypeDf, x="variableType", y="count", ax=ax)
ax.set(xlabel="特征的数据类型", ylabel="计数", title="变量的数据类型统计")
```

# 导入训练集和测试集

```python
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")


# 合并训练集和测试集
data = train.append(test)
data.reset_index(inplace=True)
data.drop("index", inplace=True, axis=1)
```

```python
data.head()
```

```python
# 整个数据集进行特征工程

data["date"] = data.datetime.apply(lambda x: x.split()[0])
data["hour"] = data.datetime.apply(lambda x: x.split()[1].split(":")[0]).astype(int)
data["year"] = data.datetime.apply(lambda x: x.split()[0].split("-")[0])
data["weekday"] = data.date.apply(lambda dateString: datetime.strptime(dateString, "%Y-%m-%d").weekday())
data["month"] = data.date.apply(lambda dateString: datetime.strptime(dateString, "%Y-%m-%d").month)
```

```python
data.describe()
```

```python
# 由于风速一定不可能为0, 但是数据集中有风速为0的数据,
# 我们用随机森林拟合那些风速不为0的数据, 并预测那些风速为0的预测数据

from sklearn.ensemble import RandomForestRegressor


dataWind0 = data[data.windspeed == 0]
dataWindNot0 = data[data.windspeed != 0]
rfr_model = RandomForestRegressor()
windColumns = ["temp", "humidity", "season", "weather", "hour"]
rfr_model.fit(dataWindNot0[windColumns], dataWindNot0["windspeed"])

wind0Values = rfr_model.predict(dataWind0[windColumns])
dataWind0["windspeed"] = wind0Values
data = dataWindNot0.append(dataWind0)
data.reset_index(inplace=True)
data.drop("index", inplace=True, axis=1)

```

```python
data.columns
```

```python
categoricalFeatures = ["holiday", "season", "weather", "workingday", "hour", "year", "weekday", "month"]
numericalFeatures = ["temp", "windspeed", "humidity"]
dropFeatures = ["casual", "registered", "atemp", "datetime", "date", "count"]


# 将类别变量转换为category数据类型
for var in categoricalFeatures:
    data[var] = data[var].astype("category")
```

```python
# 拆分训练集和测试集
train = data[pd.notnull(data["count"])].sort_values(by=["datetime"])
test = data[pd.isnull(data["count"])].sort_values(by=["datetime"])
test_datetimecol = test["datetime"]

train_y = train["count"]
train_X = train.drop(dropFeatures, axis=1)
test_X = test.drop(dropFeatures, axis=1)
```

```python
# 评分指标RMSLE
def rmsle(y_preds, y, convertExp=True):
    if convertExp:
        y_preds = np.exp(y_preds)
        y = np.exp(y)
    Log1 = np.nan_to_num(np.array([np.log(w) for w in y_preds]))
    Log2 = np.nan_to_num(np.array([np.log(v) for v in y]))
    calc =(Log1-Log2) ** 2
    return np.sqrt(np.mean(calc))
```

# 建立模型

## 普通线性回归模型

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
import warnings
pd.options.mode.chained_assignment = None
warnings.filterwarnings("ignore", category=DeprecationWarning)

# 初始化线性回归模型
linear_model = LinearRegression()

# 训练模型
train_y_log = np.log1p(train_y)  # 将目标变量值加1后进行自然对数转换
linear_model.fit(train_X, train_y_log)

# 对训练集进行预测
preds = linear_model.predict(train_X)
linear_rmsle = rmsle(preds, train_y_log)
print("普通线性回归模型的RMSLE值: ", linear_rmsle)
```

## 正则化模型-Ridge

```python
ridge_model = Ridge()
ridge_params = {
    "max_iter": [3000],
    "alpha": [0.1, 1, 2, 3, 4, 10, 30, 100, 200, 300, 400, 800, 900, 1000]
}
rmsle_scorer = metrics.make_scorer(rmsle, greater_is_better=False)
grid_ridge_model = GridSearchCV(ridge_model, ridge_params,
                               scoring=rmsle_scorer, cv=5)
train_y_log = np.log1p(train_y)
grid_ridge_model.fit(train_X, train_y_log)
preds = grid_ridge_model.predict(train_X)
ridge_rmsle = rmsle(preds, train_y_log)
print("Ridge回归的最佳参数: ", grid_ridge_model.best_params_)
print("Ridge回归的RMSLE值: ", ridge_rmsle)


fig, ax = plt.subplots()
fig.set_size_inches(12, 5)
df = pd.DataFrame(grid_ridge_model.grid_scores_)
df["alpha"] = df["parameters"].apply(lambda x: x["alpha"])
df["rmsle"] = df["mean_validation_score"].apply(lambda x: -x)
# 点线图
sns.pointplot(data=df, x="alpha", y="rmsle", ax=ax)
```

## 正则化模型-Lasso

```python
lasso_model = Lasso()
alpha = 1/np.array([0.1, 1, 3, 4, 10, 30, 100, 200, 300, 400, 800, 900, 1000])
alpha
lasso_params = {
    "max_iter": [3000],
    "alpha": alpha
}
rmsle_scorer = metrics.make_scorer(rmsle, greater_is_better=False)
grid_lasso_model = GridSearchCV(lasso_model, lasso_params, 
                               scoring=rmsle_scorer, cv=5)
train_y_log = np.log1p(train_y)
grid_lasso_model.fit(train_X, train_y_log)
preds = grid_ridge_model.predict(train_X)
lasso_rmsle = rmsle(preds, train_y_log)
print("Lasso回归的最佳参数: ", grid_lasso_model.best_params_)
print("Lasso回归的RMSLE值: ", lasso_rmsle)


fig, ax = plt.subplots()
df = pd.DataFrame(grid_lasso_model.grid_scores_)
df["alpha"] = df["parameters"].apply(lambda x: x["alpha"])
df["alpha"] = round(df["alpha"], 4)
df["rmsle"] = df["mean_validation_score"].apply(lambda x: -x)
# 点线图
sns.pointplot(data=df, x="alpha", y="rmsle", ax=ax)
plt.tight_layout()
```

## 融合模型-随机森林(RandomForest)

```python
from sklearn.ensemble import RandomForestRegressor

rfr_model = RandomForestRegressor(n_estimators=100)
train_y_log = np.log1p(train_y)
rfr_model.fit(train_X, train_y_log)
preds = rfr_model.predict(train_X)
rfr_rmsle = rmsle(preds, train_y_log)
print("随机森林回归的RMSLE值: ", rfr_rmsle)
```

## 融合模型-梯度上升(Gradient Boost)

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor(n_estimators=4000, alpha=0.01)
train_y_log = np.log1p(train_y)
gb_model.fit(train_X, train_y_log)
preds = rfr_model.predict(train_X)
gb_rmsle = rmsle(preds, train_y_log)
print("梯度上升的RMSLE值: ", gb_rmsle)
```

## xgboost

```python
import xgboost as xgb

train_y_log = np.log1p(train_y)
dtrain = xgb.DMatrix(train_X.values, label=train_y_log)
params = {
    "max_depth": 6,
    "eta": 0.1,
    "tree_num": 1000
}
model = xgb.cv(params, dtrain, num_boost_round=500, early_stopping_rounds=100)
```

```python
model.ix[30:, ["test-rmse-mean", "train-rmse-mean"]].plot()
```

```python
# 下面的模型参数是根据上述的图形看到的
xgb_model = xgb.XGBRegressor(n_estimators=250, max_depth=6, learning_rate=0.1)
xgb_model.fit(train_X.values, train_y_log)
preds = xgb_model.predict(train_X.values)
xgb_rmsle = rmsle(preds, train_y_log)
print("xgboost的RMSLE值: ", xgb_rmsle)
```

```python
df = pd.DataFrame({
        "Model": ["普通线性回归", "Ridge回归", "Lasso回归", \
                "随机森林", "梯度上升", "xgboost"],
        "RMSLE": [linear_rmsle, ridge_rmsle, lasso_rmsle, \
                 rfr_rmsle, gb_rmsle, xgb_rmsle],
    })
df.sort_values(by=["RMSLE"], ascending=True)
```

# 提交预测

```python
rfr_results = np.exp(rfr_model.predict(test_X))-1
gb_results = np.exp(gb_model.predict(test_X))-1
```

```python
results = 0.5*rfr_results + gb_results*0.5
```
