# 特征处理

## 将特征datetime列创建新的特征列

创建新的列: <strong>年, 月, 日, 时, 一周中的第几天</strong>。

```python
from datetime import *
import calendar


data_train["year"] = data_train["datetime"].apply(lambda datestring:
                            datetime.strptime(datestring,"%Y-%m-%d %H:%M:%S").year)
data_train["month"] = data_train["datetime"].apply(lambda datestring:
                            datetime.strptime(datestring,"%Y-%m-%d %H:%M:%S").month)
data_train["day"] = data_train["datetime"].apply(lambda datestring:
                            datetime.strptime(datestring,"%Y-%m-%d %H:%M:%S").day)
data_train["hour"] = data_train["datetime"].apply(lambda datestring:
                            datetime.strptime(datestring,"%Y-%m-%d %H:%M:%S").hour)
data_train["weekday"] = data_train["datetime"].apply(lambda datestring:
                            calendar.day_name[datetime.strptime(datestring,"%Y-%m-%d %H:%M:%S").weekday()])
```

## 去掉明显不需要的特征列

```python
dropVariables = ["datetime", "casual", "registered"]
data_train = data_train.drop(dropVariables, axis=1)
```

## 数据可视化

### 离群点分析

画出<strong>连续特征</strong>的箱线图

```python
continuVariable = ["temp", "atemp", "humidity", "windspeed"]
rows = [0, 0, 1, 1]
cols = [0, 1, 0, 1]

fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(12, 5)
for var, i, j in zip(continuVariable, rows, cols):
    sns.boxplot(data_train[var], orient="v", ax=axes[i][j])
axes[0][0].set(title="实际气温分布的箱线图", ylabel="实际气温")
axes[0][1].set(title="感觉气温分布的箱线图", ylabel="感觉气温")
axes[1][0].set(title="相对湿度分布的箱线图", ylabel="相对湿度")
axes[1][1].set(title="风速分布的箱线图", ylabel="风速")
plt.tight_layout()
```

从以上的<strong>箱线图</strong>可以看出各个连续特征数值的<strong>最小值、下四分位数、中位数、上四分位数、最大值</strong><br>
<strong>箱线图可以发现数据中的离群点</strong>.<br>
比如说我们可以从上图中看到相对湿度和风速分布存在异常点。<br>

```python
fig, ax = plt.subplots()

sns.boxplot(data_train["count"], orient="v")
ax.set(title="目标变量: 总的租车数量的分布箱线图", ylabel="总的租车数量")
```

箱线图的盒子两端加上1.5倍的四分位距的范围之外的点为异常值。(四分位距表示上四分位数与下四分位数的差值)<br>
如图我们发现目标变量count出现了一些离群点。<br>
通过以下操作查看连续特征的详细信息。<br>

```python
continuVariable.append("count")
data_train[continuVariable].describe()
```

从上表中, 以目标变量count为例, 我们可以得出如下信息:<br>
&emsp;总的租车数量的中位数是145, 50%的值都落在42和284之间,最小值为1,最大值为977。<br>
<br>

根据3σ原则，异常值被定义为数据值与总体数据平均值的差的绝对值超过总体数据3倍标准差的值。<br>
由于原有训练集有10886条, 去除异常值后只少了147条数据。故可直接删除数据<br>

```python
data_train_rmOutliers = data_train[np.abs(data_train["count"] - data_train["count"].mean()) <= 3*data_train["count"].std()]
```

```python
print("除去异常值之后少了{}条数据".format(data_train.shape[0] - data_train_rmOutliers.shape[0]))
```

### 类别特征对目标变量的影响

选取一些类别特征, 作这些变量与目标变量关系的<strong>箱线图</strong>

```python
categoryVariable = ["season", "holiday", "workingday", "weather", "year", "month", "hour", "weekday"]
rows = [0, 0, 1, 1, 2, 2, 3, 3]
cols = [0, 1, 0, 1, 0, 1, 0, 1]

season_mapping = {1: "春季", 2: "夏季", 3: "秋季", 4: "冬季"}
holiday_mapping = {1: "是", 0: "否"}
workingday_mapping = {1: "是", 0: "否"}
weather_mapping = {1: "晴朗多云", 2: "多雾少云", 3: "雨雪散云", 4: "大雨雷暴"}
weekday_mapping = {"Monday": "星期一", "Tuesday": "星期二", "Wednesday": "星期三",
                  "Thursday": "星期四", "Friday": "星期五", "Saturday": "星期六", "Sunday": "星期天"}

data_train_rmOutliers["season"] = data_train_rmOutliers["season"].map(season_mapping)
data_train_rmOutliers["holiday"] = data_train_rmOutliers["holiday"].map(holiday_mapping)
data_train_rmOutliers["workingday"] = data_train_rmOutliers["workingday"].map(workingday_mapping)
data_train_rmOutliers["weather"] = data_train_rmOutliers["weather"].map(weather_mapping)
data_train_rmOutliers["weekday"] = data_train_rmOutliers["weekday"].map(weekday_mapping)

fig, axes = plt.subplots(nrows=4, ncols=2)
fig.set_size_inches(12, 7)
for var, i, j in zip(categoryVariable, rows, cols):
    sns.boxplot(x=var, y="count", data=data_train_rmOutliers, orient="v", ax=axes[i][j])
    
axes[0][0].set(xlabel="季节", ylabel="租车的数量", title="不同季节租车数量的箱线图")
axes[0][1].set(xlabel="节假日", ylabel="租车的数量", title="是否是节假日租车数量的箱线图")
axes[1][0].set(xlabel="工作日", ylabel="租车的数量", title="是否是工作日租车数量的箱线图")
axes[1][1].set(xlabel="天气状况", ylabel="租车的数量", title="不同天气状况租车数量的箱线图")
axes[2][0].set(xlabel="年份", ylabel="租车的数量", title="不同年份租车数量的箱线图")
axes[2][1].set(xlabel="月份", ylabel="租车的数量", title="不同月份租车数量的箱线图")
axes[3][0].set(xlabel="时刻", ylabel="租车的数量", title="不同时刻租车数量的箱线图")
axes[3][1].set(xlabel="星期", ylabel="租车的数量", title="不同星期日租车数量的箱线图")

plt.tight_layout()
```

从上面的<strong>箱线图</strong>中可知:<br>
1, <strong>季节的影响</strong>: 夏季、秋季和冬季的租车数量大体一致且分布均匀,而春季的租车数量明显少于其他三个季节。<br>
2, <strong>工作日vs节假日的影响</strong> : 节假日或工作对租车数量的影响不是特别大。是节假日非工作日的租车数量的最大值大于是工作日非节假日。<br>
3, <strong>天气状况的影响</strong>: 可以看到不同的天气状况条件下租车数量明显存在差异。天气良好的情况比不良天气的租车数量要多很多。<br>
4, <strong>不同年份的影响</strong>: 2012年的租车数量明显多余2011年。<br>
5, <strong>不同月份的影响</strong>: 租车数量比较多的有5月到9月。这是一年中夏秋季节。<br>
6, <strong>不同时刻的影响</strong>: 明显看出一天中租车数量最多的时刻是早上8点到9点和下午5点到6。这段时间是上下班的高峰期。<br>
7, <strong>不同星期的影响</strong>：可以看出每个星期数量都差不多。只是星期一到星期五分布较为均匀。星期六和星期天明显正偏<br>
<br>

选取一些类别特征, 作这些变量与目标变量关系的<strong>直方图、折线图</strong>

```python
sort_order1 = [2012, 2011]
sort_order2 = ["星期一", "星期二", "星期三", "星期四", "星期五", "星期六", "星期天"]
sort_order3 = ["春季", "夏季", "秋季", "冬季"]

fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=2)
fig.set_size_inches(14, 7)
sns.barplot(x="month", y="count", hue="year", data=data_train_rmOutliers, ax=ax1[0])
ax1[0].set(xlabel="月份", ylabel="租车数量", title="不同年份不同月份租车数量分布的直方图")

df = pd.DataFrame(data_train_rmOutliers.groupby(["year", "month"])["count"].mean()).reset_index()
sns.pointplot(x="month", y="count", hue="year", data=df, ax=ax1[1])
ax1[1].set(xlabel="月份", ylabel="租车数量", title="不同年份不同月份租车数量分布的折线图")

sns.pointplot(x="season", y="count", hue = "year", hue_order=sort_order1, data=data_train_rmOutliers, ax=ax2[0])
ax2[0].set(xlabel="季节", ylabel="租车数量", title="不同年份不同季节的租车数量的平均值")

df1 = pd.DataFrame(data_train_rmOutliers.groupby(["year", "weather"])["count"].mean()).reset_index()
sns.pointplot(x="weather", y="count", hue="year", data=df1, ax=ax2[1])
ax2[1].set(xlabel="天气", ylabel="租车数量", title="不同年份不同天气的租车数量的平均值")

df3 = pd.DataFrame(data_train_rmOutliers.groupby(["weekday", "hour"])["count"].mean()).reset_index()
sns.pointplot(x="hour", y="count", hue="weekday", data=df3, hue_order=sort_order2, ax=ax3[0])
ax3[0].set(xlabel="时刻", ylabel="租车数量", title="不同星期不同时刻的租车数量的平均值")

df4 = pd.DataFrame(data_train_rmOutliers.groupby(["season", "hour"])["count"].mean()).reset_index()
sns.pointplot(x="hour", y="count", hue="season", data=df4, hue_order=sort_order3, ax=ax3[1])
ax3[1].set(xlabel="时刻", ylabel="租车数量", title="不同季节不同时刻的租车数量的平均值")

plt.tight_layout()
```

从以上直方图或折线图可以看到:<br>

3, 星期一到星期五租车数量较多时刻是上午8点和下午5点到6点左右。星期六和星期天租车数量较多的时刻集中在上午11点到下午4点左右。<br>

### 连续特征对目标变量的影响

#### 目标变量count分布的直方图和密度曲线图<br>
#### 目标变量count的实际分布与正态分布对比的概率图<br>

```python
from matplotlib.font_manager import FontProperties
font=FontProperties(fname=r'C:\Windows\Fonts\simhei.ttf',size=14)
sns.set(font=font.get_name())


from scipy import stats

fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(12, 5)

sns.distplot(data_train_rmOutliers["count"], kde=True, ax=axes[0][0])
axes[0][0].set(title="目标变量count分布的直方图和密度曲线图")
stats.probplot(data_train_rmOutliers["count"], dist="norm", fit=True, plot=axes[0][1])
axes[0][1].set(title="目标变量count的实际分布与正态分布对比的概率图")

sns.distplot(np.log1p(data_train_rmOutliers["count"]), kde=True, ax=axes[1][0])
axes[1][0].set(title="目标变量count取值加1再经过对数转换后的分布直方图和密度曲线图")
stats.probplot(np.log1p(data_train_rmOutliers["count"]), dist="norm", fit=True, plot=axes[1][1])
axes[1][1].set(title="目标变量count取值加1再经过对数转换后的的实际分布与正态分布对比的概率图")

plt.tight_layout()
```

```python
def get_rf_score(train_X, train_y):
    rf_model = RandomForestRegressor(n_estimators=140,max_features="sqrt")
    rf_model.fit(train_X, train_y)
    rf_model_scores = cross_val_score(rf_model, train_X, train_y, cv=10, scoring="neg_mean_squared_error")
    rf_model_score = np.sqrt(-rf_model_scores).mean()
    return rf_model_score
```
